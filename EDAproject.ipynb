# Boxplot
A boxplot (also known as a whisker plot) is a powerful visualization tool used to display the distribution of a dataset and identify potential outliers. It provides a summary of one or more continuous variables, including their median, quartiles, and potential outliers.




```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product Category', y='Total Purchase Amount',hue='Gender', data=df)
```

```python
Month_Purchase = df.groupby('month')['Total Purchase Amount'].sum()
Month_Purchase=Month_Purchase/1000
print(Month_Purchase)
```

```python
colour = sns.color_palette("husl", 12)

plt.bar(Month_Purchase.index, Month_Purchase, color=colour, edgecolor='black')

for i, value in enumerate(Month_Purchase):
    plt.text(i+1, value / 2, round(value, 2), ha='center', va='center', color='white', fontsize=12,rotation=90)

plt.xticks(ticks=range(1, 13), rotation=45, fontsize=10)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Purchase Amount', fontsize=12)
plt.title('Total Purchase Amount by Month', fontsize=14)
```

# Scatter plot

```python
new_df = df.head(300)
m = new_df['Total Purchase Amount'] / 100

np.random.seed(0)
sizes = np.random.uniform(20, 100, size=len(new_df))

plt.figure(figsize=(10, 6))
scatter = plt.scatter(x=new_df['Age'], y=m, c=m, cmap='summer', edgecolor='k', s=sizes, alpha=0.7)

plt.colorbar(scatter, label='Total Purchase Amount (normalized)')
plt.xlabel('Age', fontsize=12)
plt.ylabel('Total Purchase Amount (normalized)', fontsize=12)
plt.title('Scatterplot of Age vs Normalized Total Purchase Amount', fontsize=14)

plt.show()
```

# Conclusion

In the context of Exploratory Data Analysis (EDA) on e-commerce customer data, custom ratios can provide insightful conclusions. Hereâ€™s a general framework for summarizing such an analysis:

1. **Customer Segmentation**: By analyzing custom ratios, you might uncover different segments within your customer base, such as high-value customers versus low-value ones. This helps in targeted marketing and personalized customer service.

2. **Purchase Patterns**: Ratios like average order value to purchase frequency can reveal patterns in buying behavior, helping you understand how often and how much different customer groups spend.

3. **Customer Retention**: Ratios that involve repeat purchases or time between purchases can give insights into customer loyalty and retention rates, allowing for strategies to improve these metrics.

4. **Revenue Generation**: Custom ratios such as revenue per customer or customer lifetime value can indicate which segments are most profitable and highlight areas for potential revenue growth.

5. **Churn Analysis**: Ratios that track changes in customer behavior over time can identify signs of churn and help in developing retention strategies.

6. **Marketing Effectiveness**: Comparing ratios before and after marketing campaigns can help measure the effectiveness of those campaigns.

In summary, custom ratios in EDA can reveal key insights about customer behavior, profitability, and the effectiveness of marketing strategies, guiding data-driven decision-making for e-commerce businesses.
